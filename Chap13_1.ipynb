```python
predictors = ['Family', 'Education', 'Securities Account',
       'CD Account', 'Online', 'CreditCard', 'Age_(31.8, 40.6]',
       'Age_(40.6, 49.4]', 'Age_(49.4, 58.2]', 'Age_(58.2, 67.0]',
       'Experience_(1.6, 6.2]', 'Experience_(6.2, 10.8]',
       'Experience_(10.8, 15.4]', 'Experience_(15.4, 20.0]',
       'Experience_(20.0, 24.6]', 'Experience_(24.6, 29.2]',
       'Experience_(29.2, 33.8]', 'Experience_(33.8, 38.4]',
       'Experience_(38.4, 43.0]', 'Income_(51.2, 94.4]',
       'Income_(94.4, 137.6]', 'Income_(137.6, 180.8]',
       'Income_(180.8, 224.0]', 'CCAvg_(1.667, 3.333]', 'CCAvg_(3.333, 5.0]',
       'CCAvg_(5.0, 6.667]', 'CCAvg_(6.667, 8.333]', 'CCAvg_(8.333, 10.0]',
       'Mortgage_(63.5, 127.0]', 'Mortgage_(127.0, 190.5]',
       'Mortgage_(190.5, 254.0]', 'Mortgage_(254.0, 317.5]',
       'Mortgage_(317.5, 381.0]', 'Mortgage_(381.0, 444.5]',
       'Mortgage_(444.5, 508.0]', 'Mortgage_(508.0, 571.5]',
       'Mortgage_(571.5, 635.0]']

outcome = ['Personal Loan']

y = ub[outcome]
X = ub[predictors]

# partition data
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
#First model with Logistic regression

# fit a logistic regression (set penalty=l2 and C=1e42 to avoid regularization)
logit_reg = LogisticRegression(penalty="l2", C=1e42, solver='liblinear')
logit_reg.fit(train_X, train_y)
```

```python
logit_pred = logit_reg.predict(valid_X)
classificationSummary(valid_y, logit_pred)
```

```python
# Second model with KNN
knn = KNeighborsClassifier(n_neighbors=3).fit(train_X, train_y)
knn_pred = knn.predict(valid_X)
classificationSummary(valid_y, knn_pred)
```

```python
#Second model with classification trees
ClassTree = tree.DecisionTreeClassifier().fit(train_X, train_y)
Decision_pred = ClassTree.predict(valid_X)
classificationSummary(valid_y, Decision_pred)
```

```python
Decision_pred = pd.DataFrame(Decision_pred, index = valid_y.index)
Decision_pred
```
