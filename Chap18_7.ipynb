```python
%matplotlib inline
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import statsmodels.formula.api as sm
from statsmodels.tsa import tsatools
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
sham = pd.read_csv('ShampooSales.csv')
```

```python
sham
```

```python
#convert to time series
sham['Date'] = pd.to_datetime(sham.Month, format = '%b-%y')
sales_ts = pd.Series(sham['Shampoo Sales'].values, index=sham.Date, name = 'Sales')
sales_ts_index = pd.DatetimeIndex(sales_ts)
```

```python
fig, ax = plt.subplots(figsize=(8, 7))
ax = sales_ts.plot(ax=ax, color='red')
ax.set_ylabel('Shampoo Sales')
```
